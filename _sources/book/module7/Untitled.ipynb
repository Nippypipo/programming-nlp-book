```python
import pandas as pd
```

```python
d = pd.read_csv('data/tweet_tokenized.csv')
```

```python

docs = ['|'.join(eval(x)) if x != '' else '' for x in d.tok.to_list() ]
```

```python
d['tokens'] = docs
```

```python
d.drop(['tok', 'Unnamed: 0'], axis=1).to_csv('data/tweet_tokenized.csv', index=False)
```
